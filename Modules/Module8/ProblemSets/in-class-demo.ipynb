```r
library(sommer)
```

```r
# split plot example
# code from https://rdrr.io/cran/agridat/man/yates.oats.html 
data(DT_yatesoats)
DT <- DT_yatesoats
DT
```

```r
split.fit <- mmer(Y~ 1+ V + N + V:N,
           random = ~ B + B:MP,
           data = DT)

summary.mmer(split.fit)
```

```r
#testing to see if the variety effect is significant
wald.test(b = split.fit$Beta$Estimate, Sigma = split.fit$VarBeta, Terms=2)

#testing to see if the nitrogen effect is significant
wald.test(b = split.fit$Beta$Estimate, Sigma = split.fit$VarBeta, Terms=3)

#creating a custom contrast to test for a significant difference between Marvellous and Victory
LL <- matrix(0,nrow=1, ncol=12)
LL[1,2] <- 1
LL[1,3] <- -1
LL
wald.test(b = split.fit$Beta$Estimate, Sigma = split.fit$VarBeta, L=LL)
```

## The Wald test shows that the main effects of the two treatments are insignificant
### It is common to see spatial varaition in field experiments, so lets check for row and column effects

```r
#Checking for a spatial effect of column
res=split.fit$residuals

plot(DT$col,res)
```

```r
split.fit2 <- mmer(Y~ 1+ col+ V + N + V:N,
           random = ~ B + B:MP,
           data = DT)

summary.mmer(split.fit2)
```

```r
#Repeating the test after accounting for spatial variation
#testing to see if the variety effect is significant
wald.test(b = split.fit$Beta$Estimate, Sigma = split.fit$VarBeta, Terms=3)

#testing to see if the nitrogen effect is significant
wald.test(b = split.fit$Beta$Estimate, Sigma = split.fit$VarBeta, Terms=4)
```
